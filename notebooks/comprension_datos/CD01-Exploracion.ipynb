##### Importación de librerías

```python
import pandas as pd
# from ydata_profiling import ProfileReport
import dtale as dt
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

----

##### Lectura de los datasets

```python
df_creditos = pd.read_csv("../../data/raw/datos_creditos.csv", sep=";")
df_creditos.head(1)
```

```python
df_tarjetas = pd.read_csv("../../data/raw/datos_tarjetas.csv", sep=";")
df_tarjetas.head(1)
```

----

##### Descripción de los datasets

```python
def reporte_descripcion_dataset(df):
    columnas = df.columns
    print("Columnas del dataset:\n")
    for col in columnas:
        print(col)
    print(f"\nCantidad de filas: {df.shape[0]}")
```

```python
print("Descripción del dataset 'datos_creditos'")
reporte_descripcion_dataset(df_creditos)
```

```python
print("Descripción del dataset 'datos_tarjetas'")
reporte_descripcion_dataset(df_tarjetas)
```

----

##### Obtención de metadatos

```python
# Se usa la librería dtale para poder obtener una vista interactiva de los datos

# Dataset: datos_creditos

creditos = dt.show(df_creditos)
creditos.open_browser()
```

```python
# Dataset: datos_tarjetas

tarjetas = dt.show(df_tarjetas)
tarjetas.open_browser()
```

----

##### Gráficos de interés

```python
plt.figure(figsize=(7,5))
sns.countplot(y=df_creditos['objetivo_credito'], order=df_creditos['objetivo_credito'].value_counts().index)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.hist(df_creditos['antiguedad_empleado'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Antigüedad del Empleado (en meses)')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Antigüedad del Empleado')
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.hist(df_creditos['tasa_interes'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Tasa de Interés (%)')
plt.ylabel('Frecuencia')
plt.title('Distribución de las Tasas de Interés')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.hist(df_creditos['ingresos'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Ingresos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Ingresos')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.scatter(df_creditos['ingresos'], df_creditos['tasa_interes'], color='skyblue', alpha=0.5)
plt.xlabel('Ingresos')
plt.ylabel('Tasa de Interés (%)')
plt.title('Relación entre Ingresos y Tasa de Interés')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.countplot(y='nivel_educativo', data=df_tarjetas, palette='Set2')
plt.xlabel('Número de Clientes')
plt.ylabel('Nivel Educativo')
plt.title('Distribución del Nivel Educativo de los Clientes')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.countplot(x='personas_a_cargo', data=df_tarjetas, palette='viridis')
plt.xlabel('Personas a Cargo')
plt.ylabel('Frecuencia')
plt.title('Distribución de Personas a Cargo')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.histplot(df_tarjetas['limite_credito_tc'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Límite de Crédito')
plt.ylabel('Frecuencia')
plt.title('Distribución del Límite de Crédito')
plt.grid(True)
plt.show()
```
